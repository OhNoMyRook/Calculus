## 4.16

```python
ro = 100
L = 3
r = 1
w = 1
```

$$C=\frac{x_B}{\rho L^2}; A=\frac{x_B-r\cos \omega t}{L}; B=\frac{r \sin \omega t}{L}$$

$$\frac{dC}{dt}=-[A^2+B^2-1]; C(0)=0$$

```python
import numpy as np
```

```python
time = np.linspace(0, 100, 100000)
```

```python
def func(t_n, C_n):
    return -(((C_n*ro*L**2-r*np.cos(w*t_n))/L)**2+(r*np.sin(w*t_n)/L)**2-1)
```

```python
C = np.zeros(100000)
C[0]=0
for i in range (1, time.shape[0]):
    C[i]=(C[i-1]+(time[i]-time[i-1])*func(time[i-1],C[i-1]))
```

```python
xb = C*ro*L
ya = r*np.sin(w*time)
xa = r*np.cos(w*time)
mu = [0.01, 0.1, 0.2, 0.3, 0.5, 0.7, 0.85, 0.99]
x = []
y = []
for m in mu:
    x.append(xb[10:] - (xb[10:]-xa[10:])*m)
    y.append(ya[10:]*m)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x[0],y[0],'-',x[1],y[1],'-',x[2],y[2],'-',x[3],y[3],'-',x[4],y[4],'-',x[5],y[5],'-',x[6],y[6],'-',x[7],y[7],'-')
plt.legend(['mu=0.01','mu=0.1','mu=0.2','mu=0.3','mu=0.5','mu=0.7','mu=0.85','mu=0.99'])
plt.grid(True)
plt.title('    ')
plt.show()
```

## 6.1

```python
q1 = 1
q2 = 1
l = 5
```

```python
def f(x,y):
    return (q1*y/((x**2+y**2)**1.5)+(q2*y/(((l-x)**2+y**2)**1.5)))/((q1*x/((x**2+y**2)**1.5))-(q2*(l-x)/(((l-x)**2+y**2)**1.5)))
```

```python
y1=[]
y2=[]
xright=np.linspace(2.52,10,1000)
xleft=np.linspace(-5,2.48,1000)
start_values = np.linspace(-2,2,11)
for value in start_values:
    yr=np.zeros(xright.shape[0])
    yr[0] = value
    for j in range (1,1000):
        yr[j] = yr[j-1]+(xright[j]-xright[j-1])*f(xright[j-1], yr[j-1])

    y1.append(yr)
    y2=np.flip(y1)

x=np.concatenate((xleft, xright))
y=np.hstack((y2,y1))
```

```python
plt.plot(x,y[0],'-',x,y[1],'-',x,y[2],'-',x,y[3],'-',x,y[4],'-',x,y[5],'-',x,y[6],'-',x,y[7],'-',x,y[8],'-',x,y[9],'-',x,y[10],'-')
plt.grid(True)
plt.title('    5')
plt.show()
```
