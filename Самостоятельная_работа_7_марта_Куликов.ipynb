## Задача 1. Тринометрическая интерполяция и равномерное приближение (4 балла).

**Теоретические сведения**.

**Метод ортогонального проектирования.**

Метод ортогонального проектирования позволяет найти коэффициенты тригонометрического интерполяционного многочлена, минимизируя ошибку аппроксимации в смысле среднеквадратичного отклонения (метод наименьших квадратов в пространстве ортогональных функций).

**1. Постановка задачи**

Пусть дана функция $f(x)$, определённая на отрезке $[-1,1]$. Для интерполяции используем тригонометрический многочлен:

$$
T_m(x)=\sum_{j=0}^{m-1} a_j \cos (j t)+\sum_{j=1}^{m-1} b_j \sin (j t)
$$


где $t=\arccos x$, что соответствует переходу к равномерной сетке узлов.

**2. Определение коэффициентов через скалярное произведение**

Коэффициенты $a_j$ и $b_j$ можно найти методом ортогонального проектирования, используя скалярное произведение в пространстве функций:

$$
\langle g, h\rangle=\int_{-\pi}^\pi g(t) h(t) d t
$$

**3. Дискретное приближение интегралов**

На практике интегралы заменяются суммами по равномерной сетке узлов:

$$
\begin{aligned}
& a_j \approx \frac{2}{m} \sum_{k=0}^{m-1} f\left(x_k\right) \cos \left(j t_k\right) \\
& b_j \approx \frac{2}{m} \sum_{k=0}^{m-1} f\left(x_k\right) \sin \left(j t_k\right)
\end{aligned}
$$


где узлы $x_k$ соответствуют узлам Чебышева:

$$
x_k=\cos \left(\pi \frac{2 k-1}{2 m}\right), \quad t_k=\pi \frac{2 k-1}{2 m}
$$

**4. Итоговый интерполяционный многочлен**

Подставляя найденные коэффициенты $a_j$ и $b_j$, получаем тригонометрический интерполяционный многочлен:

$$
T_m(x)=a_0+\sum_{j=1}^{m-1}\left(a_j \cos (j t)+b_j \sin (j t)\right)
$$


где $t=\arccos x$.


**Интерполяционный многочлен в базисе многочленов Чебышева**

1. Выбор узлов интерполяции

Интерполяция в базисе Чебышевских многочленов обычно проводится в узлах Чебышева, которые минимизируют ошибку интерполяции:

$$
x_j=\cos \left(\frac{\pi(2 j+1)}{2 m}\right), \quad j=0,1, \ldots, m-1
$$

2. Вычисление значений функции

Рассчитываем значения функции в этих узлах:

$$
f_j=f\left(x_j\right)
$$


3. Разложение в ряд Чебышева

Функция $f(x)$ аппроксимируется многочленом

$$
P_m(x)=\sum_{k=0}^{m-1} c_k T_k(x)
$$


где $T_k(x)$ - Чебышевские многочлены первой степени, определяемые рекуррентным соотношением:

$$
T_0(x)=1, \quad T_1(x)=x, \quad T_{k+1}(x)=2 x T_k(x)-T_{k-1}(x)
$$


Коэффициенты $c_k$ вычисляются по формуле:

$$
c_k=\frac{2}{m} \sum_{j=0}^{m-1} f_j T_k\left(x_j\right)
$$


Для $c_0$ коэффициент делится на 2 :

$$
c_0=\frac{1}{m} \sum_{j=0}^{m-1} f_j
$$


**Интерполяционный многочлен в базисе многочленов Лежандра.**

1. Выбор узлов интерполяции

Для интерполяции можно использовать узлы Гаусса-Лежандра - корни многочлена Лежандра $P_m(x)$, или равномерную сетку в $[-1,1]$.
2. Вычисление значений функции

Рассчитываем значения функции в выбранных узлах $x_j$ :

$$
f_j=f\left(x_j\right)
$$

3. Разложение в ряд по многочленам Лежандра

Функция $f(x)$ представляется в виде линейной комбинации многочленов Лежандра:

$$
P_m(x)=\sum_{k=0}^{m-1} c_k P_k(x)
$$


где $P_k(x)$ - многочлены Лежандра, удовлетворяющие рекуррентному соотношению:

$$
\begin{gathered}
P_0(x)=1, \quad P_1(x)=x \\
P_{k+1}(x)=\frac{(2 k+1) x P_k(x)-k P_{k-1}(x)}{k+1} .
\end{gathered}
$$

4. Вычисление коэффициентов $c_k$

Коэффициенты ряда Лежандра вычисляются с помощью ортогонального разложения:

$$
c_k=\frac{2 k+1}{2} \int_{-1}^1 f(x) P_k(x) d x
$$


Если значения функции заданы в дискретных узлах, можно аппроксимировать интеграл суммированием:

$$
c_k \approx \frac{2 k+1}{2} \sum_{j=0}^{m-1} f_j P_k\left(x_j\right) w_j
$$


где $w_j$ - веса квадратурной формулы Гаусса-Лежандра.



### Задание.

**Соответствие между интерполяцией и равномерным приближением функций.**

Существует соответствие между задачей приближения функций линейными комбинациями многочленов Чебышева и тригонометрическими многочленами. Пусть на отрезке $[-1,1]$ функция $f(x)$ приближается линейными комбинациями $\sum_{j=0}^{m-1} a_j T_j(x)$. Замена переменных $x=\cos t$ сводит исходную задачу к задаче приближения функции $f(\cos t)$ линейной комбинацией $\sum_{j=0}^{m-1} a_j T_j(\cos t)=\sum_{j=0}^{m-1} a_j \cos (j t)$.

Справедливо равенство

$$
(f, g)_1=\int_{-1}^1 \frac{f(x) \bar{g}(x)}{\sqrt{1-x^2}} d x=\int_0^\pi f(\cos \theta) \bar{g}(\cos \theta) d \theta
$$

Следовательно, задача наилучшего приближения $f(x)$ в норме, соответствующей скалярному произведению $(f, g)_1$, эквивалентна задаче приближения $f(\cos \theta)$ в норме, соответствуюцей скалярному произведению $(f, g)_2=\int_0^\pi f(\cos \theta) \bar{g}(\cos \theta) d \theta$.

Точно так же существует соответствие в случае задач интерполяции и наилучшего приближения в равномерной метрике. Задача интерполирования функции многочленом по узлам $x_j=\cos \left(\pi \frac{2 j-1}{2 m}\right)$-- нулям многочлена Чебышева $T_m(x)$ - после такой замены сводится к задаче интерполирования функции $f(\cos \theta)$ при помощи тригонометрического многочлена $\sum_{j=0}^{m-1} a_j \cos (j t)$ по узлам $t_j=\pi \frac{2 j-1}{2 m}$, образующим равномерную сетку.



**Равномерная норма**

$$
\|x\|_{\mathbf{C}[a, b]}=\max _{t \in[a, b]}|x(t)|
$$


Эту норму также называют нормой Чебышёва или равномерной нормой, т.
к. сходимость по этой норме эквивалентна равномерной сходимости.
В каждом варианте дана функция: $f(x)$ и количество узлов: $m$.

В каждом варианте вам нужно использовать это соответствие. В случае, когда используется метод наименьших квадратов для интерполяции, имеется в виду, что степень многочлена должна быть выше, чем количество точек минус один.

## Решение задачи 1

```python
from scipy.special import roots_chebyt
```

```python
x = roots_chebyt(8)[0]
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
f = np.exp(x)
x, f
```

```python
difference_1 = []
difference_2 = []
difference_3 = []
difference_4 = []
difference_5 = []
difference_6 = []
difference_7 = []


for j in range(0,7):
    difference_1.append((-f[j]+f[1+j])/(x[1+j]-x[j]))
for j in range(0,6):
    difference_2.append((-difference_1[j]+difference_1[1+j])/(x[2+j]-x[j]))
for j in range(0,5):
    difference_3.append((-difference_2[j]+difference_2[1+j])/(x[3+j]-x[j]))
for j in range(0,4):
    difference_4.append((-difference_3[j]+difference_3[1+j])/(x[4+j]-x[j]))
for j in range(0,3):
    difference_5.append((-difference_4[j]+difference_4[1+j])/(x[5+j]-x[j]))
for j in range(0,2):
    difference_6.append((-difference_5[j]+difference_5[1+j])/(x[6+j]-x[j]))
difference_7.append((-difference_6[0]+difference_6[1])/(x[7]-x[0]))
```

Строим многочлен Ньютона:

```python
def Newton(t):
    return f[0]+difference_1[0]*(t-x[0])+difference_2[0]*(t-x[0])*(t-x[1])+difference_3[0]*(t-x[0])*(t-x[1])*(t-x[2])+difference_4[0]*(t-x[0])*(t-x[1])*(t-x[2])*(t-x[3])*(t-x[4])+difference_5[0]*(t-x[0])*(t-x[1])*(t-x[2])*(t-x[3])*(t-x[4])*(t-x[5])+difference_6[0]*(t-x[0])*(t-x[1])*(t-x[2])*(t-x[3])*(t-x[4])*(t-x[5])*(t-x[6])+difference_7[0]*(t-x[0])*(t-x[1])*(t-x[2])*(t-x[3])*(t-x[4])*(t-x[5])*(t-x[6])*(t-x[7])
```

```python
xarr = np.linspace(-1,1,100)
yarr = Newton(xarr)
plt.plot(x, f, 'o', xarr, yarr, '-')
plt.legend(['Исходная функция','Интерполяция многочленом Ньютона'])
plt.grid(True)
plt.show()
```

Пункты 3-4:

```python
t = np.arccos(x)
a = []
b = []
for j in range(0,8):
    coef_a = 0
    coef_b = 0
    for m in range (0,8):
        coef_a += f[m]*np.cos(j*t[m])
        coef_b += f[m]*np.sin(j*t[m])
    a.append(0.25*coef_a)
    b.append(0.25*coef_b)
```

$$
T_m(x)=\sum_{j=0}^{m-1} a_j \cos (j t)+\sum_{j=1}^{m-1} b_j \sin (j t)
$$

```python
def trig(t):
    result = 0
    for m in range (0,8):
        result += a[m]*np.cos(m*np.arccos(t))+b[m]*np.sin(m*np.arccos(t))
    return result
```

```python
y_trig = trig(xarr)
plt.plot(x, f, 'o', xarr, y_trig, '-')
plt.legend(['Исходная функция','Интерполяция тригонометрическим многочленом'])
plt.grid(True)
plt.show()
```

## Задача 2. Вычисление бесконечных сумм (2 балла).

**Дополнительный теоретический материал**

**Метод прямой сшивки.**

Метод прямой сшивки - это техника ускорения сходимости сумм бесконечных рядов, при которой часть ряда, для которой вычисление точной суммы затруднено из-за медленной сходимости, заменяется аналитически полученным приближением (асимптотическим представлением хвостовой суммы). Другими словами, вместо того чтобы суммировать все члены ряда до бесконечности, вычисляют частичную сумму первых $N$ членов, а оставшуюся «хвостовую» сумму аппроксимируют аналитически и «пришивают» (сшивают) к частичной сумме.

Основные этапы метода:
1. Вычисление частичной суммы:

Вычисляют сумму $S_N=\sum_{n=1}^N a_n$ для достаточно большого $N$, чтобы большинство вклада ряда было учтено, но чтобы вычисления были приемлемо затратными.
2. Анализ асимптотического поведения:

Исследуют асимптотику членов ряда $a_n$ при $n \rightarrow \infty$ и находят приближённое аналитическое выражение для хвостовой суммы:

$$
R_N=\sum_{n=N+1}^{\infty} a_n
$$


Это может быть выполнено, например, с использованием интегрального приближения или ряда Эйлера-Маклорена.
3. «Сшивка» результатов:

Полученную аппроксимацию хвостовой суммы $R_N$ добавляют к вычисленной частичной сумме $S_N$ :

$$
S \approx S_N+R_N
$$


Таким образом, получается улучшенное приближение к сумме бесконечного ряда, достигающее заданной точности быстрее, чем если бы сумма вычислялась напрямую.

**Метод ускорения сходимости Эйткена.**

Пусть имеется последовательность $\left\{x_n\right\}$, которая сходится к значению $L$. Метод Эйткена преобразует эту последовательность в новую, ускоренную последовательность $\left\{\hat{x}_n\right\}$ по формуле:

$$
\hat{x}_n=x_n-\frac{\left(x_{n+1}-x_n\right)^2}{x_{n+2}-2 x_{n+1}+x_n}
$$


при условии, что знаменатель $x_{n+2}-2 x_{n+1}+x_n \neq 0$.

## Решение задачи 2

```python
import numpy as np
```

```python
partial_sums = []
error_without_acc = []
error_with_acc = []
Narray = np.linspace(1, 10**7, 10**7)
```

```python
def direct_summation(N, array):  ## Частичная сумма
    result = 0.0
    for n in range(1, N + 1):
        result += (-1) ** (n + 1) / n
        array.append(result)
    return result

def aitken_acceleration(sequence, n):
    s_n = sequence[n-1]
    s_n1 = sequence[n]
    s_n2 = sequence[n+1]
    # Формула Эйткена: s = s_n - (s_{n+1} - s_n)^2 / (s_{n+2} - 2s_{n+1} + s_n)
    accelerated = s_n - ((s_n1 - s_n) ** 2) / (s_n2 - 2 * s_n1 + s_n)
    return accelerated
```

```python
direct_summation(10**7, partial_sums)
error_without_acc = partial_sums - np.log(2)
```

```python
upgraded_sequence = []
for i in range(1, np.shape(partial_sums)[0]-1):
    upgraded_sequence.append(aitken_acceleration(partial_sums,i))
for j in range(0, 10**7-2):
    error_with_acc.append(upgraded_sequence[j]-np.log(2))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(Narray[0:100], error_without_acc[0:100], '-', Narray[0:100] , error_with_acc[0:100], '-')
plt.legend(['Прямое суммирование', 'С использованием метода Эйткена'])
plt.title('Зависимость ошибки в зависимости от номера члена')
plt.grid(True)
plt.show()
```

## Задача 3. Метод плавного обрезания (3 балла).

**Теоретический материал**.

Метод плавного обрезания – это техника, при которой вместо резкого (жёсткого) обрезания или отсечения частей функции, ряда или интеграла используется специальная гладкая функция, которая постепенно уменьшает вклад этих элементов до нуля. Такой подход помогает избежать нежелательных артефактов, возникающих при резком переходе, и обеспечивает более стабильное и точное численное вычисление.

Основные идеи метода:

1. Гладкая функция отсечения:

Вместо того чтобы резко обрезать функцию (например, умножая её на характеристическую функцию, которая равна 1 в интересующем нас интервале и 0 вне его), используется функция $\chi(x)$, которая имеет следующие свойства:
- $\chi(x)=1$ на основном участке, где функция должна сохраняться;
- $\chi(x)$ плавно спадает от 1 до 0 в переходном интервале;
- $\chi(x)=0$ вне интересующего нас диапазона.

Примеры таких функций - «бамп-функции», экспоненциальные затухающие функции или функции с фиксированными гладкими переходами.

2. Применение:

Метод используется для регуляризации интегралов, сумм рядов или спектральных разложений. Например, при вычислении интеграла по бесконечному промежутку можно умножить интегрируемую функцию на функцию отсечения, которая быстро (но плавно) убывает до нуля, что позволяет ограничить область интегрирования без резкого обрезания.

3. Преимущества:
- Уменьшение эффекта «Гиббса» и других артефактов, связанных с резким отсечением.
- Сохранение гладкости аппроксимируемой функции, что важно для численного анализа и дальнейших вычислений.
- Улучшение сходимости численных методов, поскольку плавное затухание позволяет более корректно учитывать вклад удалённых элементов.

Гладкая функция отсечения также называется функцией окна (терминология из цифровой обработки сигналов, где она используется и в других целях). Примеры

1. Окно Ханна (Hann window):

Формула:

$$
w[n]=0.5\left(1-\cos \left(\frac{2 \pi n}{N-1}\right)\right), \quad n=0,1, \ldots, N-1
$$


Это окно обеспечивает плавное затухание к нулю на концах интервала.
2. Окно Хэмминга (Hamming window):

Формула:

$$
w[n]=0.54-0.46 \cos \left(\frac{2 \pi n}{N-1}\right), \quad n=0,1, \ldots, N-1
$$


Это окно снижает уровень боковых лепестков, что полезно для спектрального анализа.

3. Окно Блэкмана (Blackman window):

Формула:

$$
w[n]=0.42-0.5 \cos \left(\frac{2 \pi n}{N-1}\right)+0.08 \cos \left(\frac{4 \pi n}{N-1}\right), \quad n=0, \ldots, N-1
$$


Это окно обеспечивает еще более сильное затухание боковых лепестков по сравнению с окном Хэмминга.

4. Окно Кайзера (Kaiser window):

Формула:

$$
w[n]=\frac{I_0\left(\pi \beta \sqrt{1-\left(\frac{2 n}{N-1}-1\right)^2}\right)}{I_0(\pi \beta)}, \quad n=0,1, \ldots, N-1
$$


где $I_0$ - модифицированная функция Бесселя нулевого порядка, а $\beta$ - параметр, определяющий форму окна.
Окно Кайзера очень гибкое и позволяет регулировать компромисс между разрешением и утечкой спектра.

5. Гауссово окно (Gaussian window):

Формула:

$$
w[n]=\exp \left(-\frac{1}{2}\left(\frac{n-(N-1) / 2}{\sigma(N-1) / 2}\right)^2\right), \quad n=0, \ldots, N-1
$$

## Решение задачи 3

Простыми вычислениями на бумаги коэффициентов Фурье для данной функции (по синусам, так как нечетная), получим $b_n=\frac{4}{\pi n}, n=2k+1; b_n = 0, n=2k$

```python
def sgn(x):
    return np.where(x < 0, -1, 1)

def fourier_coefficients(N):
    b = np.zeros(N)
    for n in range(1, N + 1):
        if n % 2 == 1:
            b[n - 1] = 4 / (np.pi * n)
    return b

# Ряд Фурье с резким обрезанием
def fourier_series(x, b, N):
    result = 0.0
    for n in range(1, N + 1):
        result += b[n - 1] * np.sin(n * x)
    return result

# Окно Гаусса
def gaussian_window(n, sigma):
    return np.exp(-0.5 * (n / sigma) ** 2)

N = 100  # Количество членов ряда
sigma = 50

b = fourier_coefficients(N)

x = np.linspace(-np.pi, np.pi, 1000)

y_sharp = fourier_series(x, b, N)

b_gaussian = b * np.array([gaussian_window(n, sigma) for n in range(0, N)])
y_gaussian = fourier_series(x, b_gaussian, N)

# Графики
plt.figure(figsize=(12, 8))
plt.plot(x, sgn(x), label="Исходная функция", linewidth=2, color="black")
plt.plot(x, y_sharp, label=f"Резкое обрезание (N={N})", linestyle="--")
plt.plot(x, y_gaussian, label=f"Окно Гаусса (N={N}, σ={sigma})", linestyle="-")
plt.title("Фурье-разложение функции sgn(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```

Окно Гаусса позволило смягчить паразитные осцилляции вблизи точки разрыва и на краях интервала (эффект Гиббса)

Примечание: При $x \rightarrow 0 \cosh (x)-1 \approx \frac{x^2}{2}$ и $\cos (x)-1 \approx-\frac{x^2}{2}$, поэтому обе ветви имеют линейное приближение $y(x) \approx 5+x$.


## Решение задачи 4

По сути ищем значение производной функции y(x) в нуле.

```python
def y1(x):
    return np.cosh(x)+4+x

def y2(x):
    return 4+np.cos(x)+x

def compute_limit(delta_x):
    x_k = delta_x / 2  
    y_k = y1(x_k)       
    y_minus_k = y2(-x_k) 
    L = (y_k - y_minus_k) / (2 * x_k)  # Аппроксимация предела через конечно разностную схему 
    return L

delta_x_values = np.linspace(1e-9, 1e-6, 10**6)
results = []
errors = []

for delta_x in delta_x_values:
    L = compute_limit(delta_x)
    if np.abs(L-1)<1e-12:
        print(delta_x)
    results.append(L)
    errors.append(L-1)

plt.plot(delta_x_values, errors)
plt.grid(True)
plt.xlabel('delta')
plt.ylabel('error')
plt.title('Зависимость ошибки вычисления предела от величины delta')
plt.show()
```
