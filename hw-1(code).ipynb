```python
k = int(input())
l = int(input())
X = np.random.randint(-5, 5, (k,l))
Y = np.random.randint(-5, 5, (k,l))
print(X)
print(Y)
```

## 1.5

```python
n = int(input())
m = int(input())
X_new = np.random.randint(-5, 5, (n,m))
Y_new = np.random.randint(-5, 5, (n,m))
print(X_new)
print(Y_new)
```

```python
X_new[:,0], Y_new[:,1]
```

```python
X_new[:,0]@Y_new[:,1]
```

```python
Z = np.empty((m,m)).astype(int)
for i in range (0, m):
    for j in range (0, m):
        Z[i,j]=X_new[:,i]@Y_new[:,j]
Z
```
